# A/B Testing on generated AD prompts provided by AdSmart

A/B testing is a powerful tool that businesses can use to improve the performance of their advertisements and increase engagement with their audience. The purpose of A/B testing on the generated advertisements is to determine which version of the advertisement is more effective at achieving the desired outcome of increasing click-through rate and engagement.

In this experiment, each version of the advertisement has been crafted with a different tone, and these variations will be randomly shown to different groups of users. The performance of each version will be measured and compared against a specific metric, which is the click-through rate (CTR) in this case. By using Bayesian A/B testing with an Epsilon-Greedy algorithm, businesses can gain valuable insights into which version of their advertisement is the most effective at engaging their audience and driving clicks.

The Epsilon-Greedy algorithm is a popular approach to conducting Bayesian A/B testing that balances the exploration of new variants with the exploitation of existing knowledge. By randomly choosing a variant to test (exploration) a certain percentage of the time, and choosing the best-performing variant (exploitation) the rest of the time, the algorithm can quickly converge on the optimal variant.

This is the demonstration and simulation of Bayesian A/B Testing using Epsilon-Greedy Algorithm to show how anyone can run A/B testing to understand which of the generated three ADs provides the most click-through rate for their business.

### Hypothesis

One of the three advertisements will perform better than the other two.
